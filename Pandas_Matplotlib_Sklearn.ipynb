# Matplotlib

```python
x = np.arange(10)
y = np.random.randint(1, 10, size=10)

# Размер полотна
plt.figure(figsize=(10, 7))

# Линейная диаграмма
plt.plot(x, y)

# Заголовок
plt.title('График зависимости x от y')

# Подписи к осям
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Точки, не соединенные линиями - Scatter Plot
plt.figure(figsize=(10, 7))

plt.scatter(x, y)

plt.title('Scatter Plot')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Несколько зависимостей на одном графике

x = np.arange(10)
y = np.random.randint(1, 10, size=10)
z = np.random.randint(1, 10, size=10)

plt.figure(figsize=(10, 7))

plt.plot(x, y, color='blue', label='y')
plt.plot(x, z, color='red', label='z')

# Легенда
plt.legend()

plt.title('Данные vs Время')
plt.xlabel('Время')
plt.ylabel('Данные')

# Ограничение осей
plt.xlim(0, 10)
plt.ylim(0, 10)

# Сетка
plt.grid()

plt.show()
```

Заполним ячейки Nan с помощью статистических метрик - мод, медиан или средних.

Возьмем категориальный признак Embarked и выведем, сколько раз в этом столбце встречаются разные значения с помощью метода .value_counts:

```python
df['Embarked'].value_counts()
```

Нарисуем столбчатую диаграмму признака Embarked

```python
data = df['Embarked'].value_counts()

plt.bar(data.index, data.values)
```

Самое часто встречающееся значение - мода - выводится с помощью метода .mode():

```python
df['Embarked'].mode()
```

Возьмем колонку Age и выведем ее медиану и среднее:

```python
df['Age'].dropna().values  # возвращает array со всеми значениями колонки
```

```python
mean_Age = np.mean(df['Age'].dropna().values)  # для этого удаляем NaN

median_Age = np.median(df['Age'].dropna().values)
# Аналогично
median_Age = np.quantile(df['Age'].dropna().values, q=0.5)  # делит отсортированный массив данных в зависимости от q

print(f'Среднее = {mean_Age}')
print(f'Медиана = {median_Age}')
```

Нарисуем распределение признака Age с его средним и медианой

```python
plt.hist(df['Age'])

plt.axvline(x=mean_Age, color='r', linestyle='-.', label='mean')  # вертикальная линия по оси X
plt.axvline(x=median_Age, color='orange', linestyle='--', label='median')
# plt.axhline(y=median_Age, color='orange', linestyle='--', label='median')  # пример горизонтальной линии

plt.legend()
plt.show()
```
