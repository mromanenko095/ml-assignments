```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# x = np.random.rand(50)  # просто рандомные 50 чисел
# x = np.random.randn(50)  # числа с нормальным распределением
# x = np.random.randint(0, 2, size=50)  # ЦЕЛЫЕ числа из диапазона (здесь 0 и 1)
# x = np.arange(0, 10, 0.5)  # числа с шагом 0.5
# x = np.logspace(1, 3, 5)  # [10., 31.6, 100., 316.2, 1000.]

x = np.linspace(0, 10, 50)  # 100 чисел от 0 до 10 равномерно
y = np.ones(50) * 42

plt.figure(figsize=(10,7))

plt.plot(x, y, color='red')

plt.title("Зависимость y = y(x)")

plt.xlabel('x')
plt.ylabel('y')

plt.grid()

plt.show()

plt.savefig('зависимость.png')
```
